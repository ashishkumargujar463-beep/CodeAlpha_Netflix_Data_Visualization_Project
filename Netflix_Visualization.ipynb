## Objective

The objective of this project is to transform raw Netflix data into meaningful visualizations to identify trends, patterns, and insights for better decision-making.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv('netflix_titles.csv')

# Display first 5 rows
df.head()
```

## Visualization 1: Distribution of Movies and TV Shows

This visualization shows the proportion of Movies and TV Shows available on Netflix.

```python
plt.figure(figsize=(8,5))

sns.countplot(x='type', data=df)

plt.title('Distribution of Movies and TV Shows on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Count')

plt.show()
```

## Visualization 2: Top 10 Countries Producing Netflix Content

This visualization highlights the countries contributing the most content to Netflix.

```python
plt.figure(figsize=(12,6))

top_countries = df['country'].value_counts().head(10)

sns.barplot(x=top_countries.index,
            y=top_countries.values)

plt.xticks(rotation=45)

plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')

plt.show()
```

## Visualization 3: Netflix Content Release Trend

This visualization shows how Netflix content production has changed over the years.

```python
plt.figure(figsize=(12,6))

sns.histplot(df['release_year'], bins=30)

plt.title('Content Release Trend Over Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')

plt.show()
```

## Visualization 4: Distribution of Content Ratings

This visualization shows the distribution of ratings assigned to Netflix content.

```python
plt.figure(figsize=(12,6))

rating_counts = df['rating'].value_counts().head(10)

sns.barplot(x=rating_counts.index,
            y=rating_counts.values)

plt.xticks(rotation=45)

plt.title('Top 10 Content Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Count')

plt.show()
```

## Visualization 5: Most Popular Genres

This visualization highlights the most common genres available on Netflix.

```python
genres = df['listed_in'].str.split(', ', expand=True).stack()

top_genres = genres.value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(x=top_genres.values,
            y=top_genres.index)

plt.title('Top 10 Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')

plt.show()
```

# Key Insights

1. Movies are more prevalent on Netflix than TV Shows.

2. The United States contributes the largest share of Netflix content.

3. Netflix content production has grown rapidly after 2010.

4. TV-MA is the most common content rating on Netflix.
